### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# drop nan
df_cleaned = df.dropna(subset=['DatabaseWantToWorkWith'])
# split databases manually 
records = []
for idx, row in df_cleaned.iterrows():
    databases = str(row['DatabaseWantToWorkWith']).split(';')  # Assume they are separated by ';'
    for db in databases:
        db = db.strip()
        if db:
            records.append(db)
# new df
df_databases = pd.DataFrame(records, columns=['Database'])
# count frequency
database_counts = df_databases['Database'].value_counts()
# top 5 databases
top5 = database_counts.head(5)
# plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top5.values,
    labels=top5.index,
    autopct='%1.1f%%',   # Show percentages
    startangle=140,      # Rotate the start
    colors=plt.cm.Paired.colors  # Color map
)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Make the pie circular
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# nans
df_cleaned = df.dropna(subset=['DevType'])
# split multiple DevTypes manually into individual rows
records = []
for idx, row in df_cleaned.iterrows():
    devtypes = str(row['DevType']).split(';')  # Assume DevTypes are separated by ';'
    for dev in devtypes:
        dev = dev.strip()
        if dev:
            records.append(dev)
# new df
df_devtypes = pd.DataFrame(records, columns=['DevType'])
# count frequency
devtype_counts = df_devtypes['DevType'].value_counts()
# top 5 developer types
top5_devtypes = devtype_counts.head(5)
# plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top5_devtypes.values,
    labels=top5_devtypes.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors  # Nice distinct colors
)
plt.title('Top 5 Developer Roles')
plt.axis('equal')  # Make the pie a circle
plt.show()
```
