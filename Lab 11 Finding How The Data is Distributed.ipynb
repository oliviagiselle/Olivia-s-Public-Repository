### Step 4: Analyze Key Columns

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import matplotlib.pyplot as plt
job_sat_counts = df['JobSat'].value_counts()
plt.figure(figsize=(20, 10))
job_sat_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange', 'salmon', 'red'])
plt.title('Job Satisfaction Distribution', fontsize=14)
plt.ylabel('', fontsize=12) 
for text in plt.gca().texts:  
    text.set_fontsize(8)  
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
import seaborn as sns
# Ensure that the necessary columns are loaded properly
have_worked_with = df['LanguageHaveWorkedWith'].dropna().str.split(';', expand=True).stack()
want_to_work_with = df['LanguageWantToWorkWith'].dropna().str.split(';', expand=True).stack()

# Count the frequency of languages in both columns
have_worked_with_counts = have_worked_with.value_counts()
want_to_work_with_counts = want_to_work_with.value_counts()

# Create a DataFrame to merge the counts into one table
language_comparison = pd.DataFrame({
    'Have Worked With': have_worked_with_counts,
    'Want To Work With': want_to_work_with_counts
}).fillna(0)
language_comparison = language_comparison.sort_values(by='Have Worked With', ascending=False)
# Plotting the grouped bar chart
plt.figure(figsize=(20, 12))
language_comparison.plot(kind='bar', width=0.8, color=['skyblue', 'lightgreen'])

plt.title("Comparison of Languages Worked With vs. Want To Work With", fontsize=12)
plt.xlabel("Programming Languages", fontsize=8)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45, fontsize=6, ha='right')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
remote_work_by_region = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)
remote_work_by_region['Total'] = remote_work_by_region.sum(axis=1)
remote_work_by_region = remote_work_by_region.sort_values('Total', ascending=False)
plt.figure(figsize=(30, 10))  
remote_work_by_region.plot(kind='bar', stacked=False, color=['skyblue', 'lightgreen'], width=0.8)
plt.title("Remote Work Distribution by Region", fontsize=16)
plt.xlabel("Region (Country)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=60, ha='right', fontsize=6)
plt.tight_layout() 
plt.show()
```
